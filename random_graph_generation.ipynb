```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
# draw graph
def draw_graph(graph, node_size, node_color):
    
    pos = graphviz_layout(graph, prog="neato", args="")
    
    plt.figure(figsize=(8, 8))
    nx.draw(graph, pos, node_size=node_size, alpha=0.5, node_color=node_color)
    plt.axis("equal")
    plt.show()
    plt.close()
    
```

```python
# graph setting
node_size = 300
node_color = "purple"
seed = 1000
```

### Erdos-Renyi (ER)
- 모든 엣지가 생길 확률이 일정한 그래프 생성 모델

```python
# create graph
ER_G = nx.erdos_renyi_graph(n=20, p=0.1, seed=seed, directed=True)

# draw graph
draw_graph(graph=ER_G, node_size=node_size, node_color=node_color)
```

### Barabasi-Albert (BA)
- 임의의 규모가 없는 네트워크를 생성하는 그래프 생성 모델
- internet, social network 등을 표현할 수 있음

```python
# create graph
BA_G = nx.barabasi_albert_graph(n=100, m=2, seed=seed, initial_graph=None)

# draw graph
draw_graph(graph=BA_G, node_size=node_size, node_color=node_color)
```

### Stochastic-block (SBM)
- 특정 노드끼리 community structure을 가지는 그래프를 만들 수 있는 그래프 생성 모델
- community sturcture : 서로 가까운 관계를 가지는 노드끼리 community를 형성하는 구조
- 노드를 임의 크기의 블록으로 분할하고 블록에 따라 달라지는 확률로 노드 쌍 사이에 독립적으로 edge를 배치
- $\gamma$ 개의 block : $C_1 ... C_\gamma$
- 각 노드가 $C_i$ 에 속할 확률 $: p_i$
- $C_i$ 에 속한 노드와 $C_j$ 에 속한 노드 사이에 edge가 있을 확률 : $C[i,j]$

```python
# create graph
sizes = [9, 5, 25, 30, 30]
probs = [[0.01, 0.02, 0.05, 0.04, 0.04], [0.04, 0.02, 0.06, 0.04, 0.02], [0.01, 0.04, 0.05, 0.04, 0.02], [0.04, 0.01, 0.05, 0.04, 0.02], [0.05, 0.01, 0.03, 0.04, 0.03]]
SBM_G = nx.stochastic_block_model(sizes=sizes, p=probs, seed=seed, directed=True, selfloops=False, sparse=True)

# draw graph
draw_graph(graph=SBM_G, node_size=50, node_color=node_color)
```

### Watts-Strogatz (WS)
- 짧은 평균 경로 길이 및 높은 클러스터링을 포함하여 small-world properties를 포함하는 그래프 생성 모델

```python
# create graph
WS_G = nx.watts_strogatz_graph(n=80, k=3, p=0.3, seed=seed)

# draw graph
draw_graph(graph=WS_G, node_size=node_size, node_color=node_color)
```

### Random Regular (RR)
- 모든 꼭짓점이 동일한 수의 이웃을 가지는 그래프 (모든 꼭짓점이 같은 차수를 가짐)

```python
# create graph
RR_G = nx.random_regular_graph(d=5, n=20, seed=seed)

# draw graph
draw_graph(graph=RR_G, node_size=node_size, node_color=node_color)
```
